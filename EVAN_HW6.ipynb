## Generate Cities List

```python
# Data on world cities lat/long - retrieved 6/30/19 from https://simplemaps.com/data/world-cities
cities_data = pd.read_csv("worldcities.csv")

#Preview the data.
cities_data.head(10)
```

```python
# Randomly select 500 cities for weather data.

sample_cities = cities_data.sample(n=500)

#Visualize the selected cities DataFrame.

sample_cities.head(10)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create Counter

row_count = 0

# Create blank columns for the necessary fields.

sample_cities["Temperature"] = ""
sample_cities["Humidity"] = ""
sample_cities["Cloudiness"] = ""
sample_cities["Wind Speed"] = ""

# Loop through and get the weather data using the Open Weather API.

for index, row in sample_cities.iterrows():
    #Sleep to help with the call limit
    
    time.sleep(1)
    
    #Open weather url.
    
    url = "https://api.openweathermap.org/data/2.5/weather?"
    units = "Imperial"
    
    query_url = url + "lat="+ str(row["lat"]) + "&lon=" + str(row["lng"]) + "&appid=" + api_key + "&units=" + units
    #Print log to ensure the loop is working correctly.
    
    print("Now retieving city # " + str(row_count))
    print(query_url)
    row_count += 1
    
    #Run requests to grab the JSON at the requested URL
    
    ej_weather = requests.get(query_url).json()

    #Append the weather data to the appropriate columns.
    #Use try/except to skip any cities with errors.
    try:
        temperature = ej_weather["main"]["temp"]
        humidity = ej_weather["main"]["humidity"]
        cloudiness = ej_weather["clouds"]["all"]
        wind_speed = ej_weather["wind"]["speed"]
        
        sample_cities.set_value(index, "Temperature", temperature)
        sample_cities.set_value(index,"Humidity", humidity)
        sample_cities.set_value(index,"Cloudiness", cloudiness)
        sample_cities.set_value(index,"Wind Speed", wind_speed)
    except:
        print("Error with weather data, skipping.")
        
sample_cities
```
